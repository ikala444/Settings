# 1.1 –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –æ–±—É—á–µ–Ω–∏—è:

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ñ–∞–π–ª–æ–≤–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä–æ–π
import os

# –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ—á–Ω—ã–π –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä, –≥–¥–µ –±—É–¥—É—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –≤—Å–µ –∏–æ–±—Ä–∞–∂–µ–Ω–∏—è
images = [os.path.join('data', x) for x in os.listdir('data') if x.endswith('.bmp')]
# –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ—á–Ω—ã–π –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä, –≥–¥–µ –±—É–¥—É —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –≤—Å–µ –∞–Ω–Ω–æ—Ç–∞—Ü–∏–∏
labels = [os.path.join('data', x) for x in os.listdir('data') if x.endswith('.txt')]

# –°–æ—Ä—Ç–∏—É—Ä–µ–º —Å–æ–∑–¥–∞–Ω–Ω—ã–µ —Å–ø–∏—Å–∫–∏
images.sort()
labels.sort()
```

```python
# –ò–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ —Å –≥–æ—Ç–æ–≤—ã–º–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞–º–∏–º –ø–æ–¥ –º–∞—à–∏–Ω–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –º–æ–¥—É–ª—å –ª—è —Ä–∞–±–∏–µ–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏
from sklearn.model_selection import train_test_split
# –ò–º–ø–æ—Ä—Ç–∏—É—Ä–µ–º –±–∏–±–∏–ª–∏–æ—Ç–µ–∫—É –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –º–∞—Å—Å–∏–≤–∞–º
import numpy as np

# –†–∞–∑–±–∏–≤–∞–µ–º –Ω–∞—à—É –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É –Ω–∞ train –∏ val
train_images, val_images, train_labels, val_labels = train_test_split(images, labels, test_size=0.2, random_state=42)
# –†–∞–∑–±–∏–≤–∞–µ–º –≤—ã–±–æ—Ä–∫—É val –ø–æ–ø–æ–ª–∞–º –Ω–∞ test –∏ val
val_images, test_images, val_labels, test_labels = train_test_split(val_images, val_labels, test_size=0.5, random_state=42)

# –í—ã–≤–æ–¥–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø–æ–ª—É—á–∏–≤—à–∏—Ö—Å—è –≤—ã–±–æ—Ä–∫–∞—Ö train, val, test:
print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ train –≤—ã–±–æ—Ä–∫–µ: {np.array(train_images).size}')
print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ val –≤—ã–±–æ—Ä–∫–µ: {np.array(val_images).size}')
print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ test –≤—ã–±–æ—Ä–∫–µ: {np.array(test_images).size}')
```

```python
# –°–æ–∑–¥–∞–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é, –≤ –∫–æ—Ç—Ä–æ–π –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–∞–ø–∫–∏ –Ω–∞—â–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
dataset= 'dataset'
# –°–æ–∑–∞–µ–º –∫–æ—Ä—Ç–µ–∂, –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ–º —Ö—Ä–∞–Ω–∏—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è –≤—Å–µ—Ö –ø—É—Ç–µ–π –Ω–∞—à–µ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
dataset_dirs = ('images/train', 'images/val', 'images/test', 'labels/train', 'labels/val', 'labels/test')
# –°–æ–∑–¥–∞–µ–º —Ü–∏–∫–ª, –ø—Ä–æ–≤–µ—Ä—è—é—â–∏–π –Ω–∞–ª–∏—á–∏–µ –¥–∞–Ω–Ω—ã—Ö –ø–∞–ø–æ–∫, –ø—Ä–∏ –∏—Ö –æ—Ç—Å—É—Å—Ç–≤–∏–µ –æ–Ω –∏—Ö —Å–æ–∑–¥–∞–µ—Ç
for dir in dataset_dirs:
    # –°–æ–∑–¥–∞–µ–º –ø—É—Ç—å –¥–æ –ø–∞–ø–∫–∏
    path = os.path.join(dataset, dir)
    # –ï—Å–ª–∏ —Ç–∞–∫–æ–≥–æ –ø—É—Ç–∏ –Ω–µ—Ç, —Ç–æ —Å–æ–∑–¥–∞–µ–º –µ–≥–æ
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É –¥–ª—è –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ–∞–π–ª–æ–≤
import shutil

# –°–æ–∑–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ–∞–π–ª–æ–≤ –∏–∑ –æ–¥–Ω–æ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ –≤ –¥—Ä—É–≥—É—é
def move_files_to_folder(list_of_files, destination_folder):
    # –î–ª—è –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤ –∏–∑ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
    for file in list_of_files:
        # –ü—ã—Ç–∞–µ–º—Å—è –∫–æ–ø–∏—Ä–æ–≤–∞—Ç—å, –µ—Å–ª–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –æ—à–∏–±–∫–∞, —Ç–æ –≤—ã–≤–æ–¥–∏—Ç—Å—è –≤ –∫–æ–Ω—Å–æ–ª—å: "–û—à–∏–±–∫–∞ –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è: –∏–º—è_—Ñ–∞–π–ª–∞.—Ä–∞—Å—à–∏—Ä–µ–Ω–∏–µ"
        try:
            shutil.copy(file, destination_folder)
        except:
            print(f'–û—à–∏–±–∫–∞ –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è: {file}')

# –°–æ–∑–¥–∞–µ–º –º–Ω–æ–≥–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤ —Å–æ –≤—Å–µ–º–∏ —Ñ–∞–π–ª–∞–º–∏ –≤–Ω—Ç—É—Ä–∏
all_the_files = [[train_images, val_images, test_images], [train_labels, val_labels, test_labels]]
# –°–æ–∑–¥–∞–µ–º –∫–æ—Ä—Ç–µ–∂ —Å–æ–≤—Å–µ–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ –Ω–∞—à–µ–π –≤—ã–±–æ—Ä–∫–∏
variabels_dataset = ('images', 'labels')
# –°–æ–∑–¥–∞–µ–º –º–∞—Å—Å–∏–≤ —Å–æ –≤—Å–µ–º–∏ –∑–∞–¥–∞—á–∞–º–∏
tasks = ['train', 'val', 'test']

# –°–æ–∑–¥–∞–µ–º —Ü–∏–∫–ª, –∏—Å–ø–æ–ª—å–∑—É—é—â–∏–π –Ω–∞–ø–∏—Å–∞–Ω–Ω—É—é —Ä–∞–Ω–µ–µ —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ–∞–π–ª–æ–≤
for i in range(len(variabels_dataset)):
    path = os.path.join(dataset, variabels_dataset[i])
    for j in range(len(tasks)):
        move_files_to_folder(all_the_files[i][j], os.path.join(path, tasks[j]))
```

#### 1.2.2 –ü–µ—Ä–µ–¥–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Å–æ–≤ –≤ —Ñ–∞–π–ª–µ

# 1.3 –°–æ–∑–¥–∞–µ–º –º–æ–¥–µ–ª—å –¥–ª—è –æ–±—É—á–µ–Ω–∏—è:

```python
# –ò–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫—É ultralytics –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –º–æ–¥—É–ª—å YOLO, –≤ –∫–æ—Ç–æ—Ä–æ–º —Ö—Ä–∞–Ω—è—Ç—Å—è –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–µ –º–æ–¥–µ–ª–∏ —Ç–∏–ø–∞ yolo
from ultralytics import YOLO

# –°–æ–∑–¥–∞–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é, –≤ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å—Å—è –Ω–∞—à–∞ –º–æ–¥–µ–ª—å yolov8n, –≤–µ—Ä—Å–∏—è - 8, —Ä–∞–∑–º–µ—Ä - nano
model = YOLO('yolov8n.pt')
```

```python
# –ù–∞—á–∏–Ω–∞–µ–º –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model.train(data='data.yaml', epochs=5, optimizer='AdamW', name='yolo')
```
